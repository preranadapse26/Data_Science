```python
# What is Simple Linear Regression?
#simple linear regeression establish a linear relation between to variable, and try to  determine the strength and characteristics of the relation.
# Simple linear regresssion is algorithm used to predict the unknown value means the dependent variable using the independent variable.
# simple linear regerssion only contains two variable one dependent and one independent. With the help of independent variable we try to predict the 
# value of dependent variable.
```

```python
# What are the key assumptions of Simple Linear Regression?
# --1)linearity : Indendent variable(x) and dependent variable(y) should have linear relationship.
# 2)Independence: The observation are independent of each other.
# 3)Homoscedasticity : Also known as constant variance this assumption means the variance of the error are constant
# 4)Normality of error: Error should be normally distributed
# 5)The features should not be related or should have least relation.
```

```python
# Question 3: What is heteroscedasticity, and why is it important to address in regression
# models?
# -- heteroscedasticity means not having constant variance in data that means you a pattern in your data.
# It is important to address in model so that we know what pattern we have in data and how we can work on it.
```

```python
#  What is Multiple Linear Regression?
# multiple linear regression is extended version of simple linear regression. In multiple linear regression we predict the one dependent variable 
# by using more than one independent variable. We determine the strenght and characteristics of the relation with multiple independent variable.
```

```python
#  What is polynomial regression, and how does it differ from linear regression?
# polynomial regression means having n degree polynomial variable as a independent variable. In polynomial regression we have a higher degree
# of polynomial which do which have polynomial relation with dependent variable.
# linear regression only describe the linear relation between variables where in one unit change in x , y also change by one unit.while
# polynomial regression change n degree polynomial.
```

```python
# Implement a Python program to fit a Simple Linear Regression model to
# the following sample data:
# ● X = [1, 2, 3, 4, 5]
# ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
# Plot the regression line over the data points
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape X for scikit-learn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(X,Y, test_size = 0.2)
from sklearn.linear_model import LinearRegression
```

```python
x_train
```

```python
model =LinearRegression()
```

```python
model.fit(x_train, y_train)
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
plt.scatter(X, Y, color='black', label = "Actual")
plt.plot(X, y_pred, color ='blue',linewidth =3, label="Predict linear regression line")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
